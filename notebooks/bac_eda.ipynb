```python
# Calculating moving averages
df['SMA_20'] = df['Close'].rolling(window=20).mean()
df['SMA_100'] = df['Close'].rolling(window=100).mean()
df['SMA_200'] = df['Close'].rolling(window=200).mean()

# Plotting the closing price and moving averages
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Close Price', color='blue')
plt.plot(df.index, df['SMA_20'], label='20-day SMA', color='orange')
plt.plot(df.index, df['SMA_100'], label='100-day SMA', color='green')
plt.plot(df.index, df['SMA_200'], label='200-day SMA', color='red')

# Plotting th egraph
plt.legend()
plt.title('BAC Close Price with 20, 100, and 200-day Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df[['Open', 'High', 'Low', 'Close', 'Volume']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Plotting the graph on Daily Returns
df['Daily_Return'] = df['Close'].pct_change()
plt.plot(df['Date'], df['Daily_Return'])
plt.title('Daily Returns Over Time for BAC')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.show()
```

```python
#Plotting the graph on Close Price Distribution
plt.hist(df['Close'], bins=30, edgecolor='black')
plt.title('Close Price Distribution for BAC')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Plotting the graph on Volatility
df['Volatility'] = df['Close'].rolling(window=20).std()
plt.plot(df['Date'], df['Volatility'])
plt.title('Price Volatility (20-day Rolling Std)')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.show()
```

```python
#Plotting the graph between Close price and Volume
plt.scatter(df['Close'], df['Volume'], alpha=0.5)
plt.title('Close Price vs Volume for BAC')
plt.xlabel('Close Price')
plt.ylabel('Volume')
plt.show()
```

```python
import statsmodels.api as sm

# Performing seasonal decomposition on the 'Close' prices
decomposition = sm.tsa.seasonal_decompose(bac_data['Close'], model='additive', period=365)  # Assuming yearly seasonality

# Plotting the decomposition
plt.figure(figsize=(14, 10))

# Plotting Original series
plt.subplot(4, 1, 1)
plt.plot(bac_data['Close'], label='Original', color='black')
plt.legend(loc='upper left')
plt.title('Time Series Decomposition for BAC - Close Price')

# Plotting Trend
plt.subplot(4, 1, 2)
plt.plot(decomposition.trend, label='Trend', color='blue')
plt.legend(loc='upper left')

# Plotting Seasonality
plt.subplot(4, 1, 3)
plt.plot(decomposition.seasonal, label='Seasonality', color='green')
plt.legend(loc='upper left')

# Plotting Residuals
plt.subplot(4, 1, 4)
plt.plot(decomposition.resid, label='Residuals', color='red')
plt.legend(loc='upper left')

# Displaying the output
plt.tight_layout()
plt.show()
```
