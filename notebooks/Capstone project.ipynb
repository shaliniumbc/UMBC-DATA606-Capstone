# Concatinating all the dataframes into one using glob

```python
import glob
import pandas as pd

# Using glob to find all CSV files
file_paths = glob.glob(r'/Users/shalinimopuri/Desktop/NIFTY50 Project/extracted_files/*.csv')

# Creating an empty list to hold the DataFrames
df_list = []

# Iterating through the file paths and reading each CSV into a DataFrame
for file_path in file_paths:
    df = pd.read_csv(file_path)
    df_list.append(df)

# Concatenate all the DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

# Saving the combined DataFrame to a new CSV file
combined_df.to_csv('combined_csv_file.csv', index=False)
print(combined_df.head())
```

```python
dfs = []
for file_path in file_paths:
    # Avoiding reading metadata file
    if 'stock_metadata' not in file_path and 'NIFTY50_all' not in file_path:
        df = pd.read_csv(file_path, parse_dates=['Date'])
        df['Stock'] = file_path.split('/')[-1].replace('.csv', '')  # Adding stock name column
        dfs.append(df)
```

```python
Nifty_50 = pd.concat(dfs, ignore_index=True)
Nifty_50.head()
```

```python
# Distribution of Trades
plt.figure(figsize=(10, 6))
sns.histplot(Nifty_50['Trades'], kde=True, bins=30)
plt.title('Distribution of Trades')
plt.xlabel('Number of Trades')
plt.ylabel('Frequency')
plt.show()
```

```python
#Top 5 Companies by Sum of Trades
since_inception_ranking = Nifty_50.groupby('Symbol').agg({'Trades': ['mean', 'sum']}).sort_values(by=('Trades', 'sum'), ascending=False).head(5)

# Renaming the columns
since_inception_ranking.columns = ['Mean Trades', 'Sum Trades']
```

```python
# Plotting top 5 Companies by Sum of Trades
plt.figure(figsize=(10, 6))
since_inception_ranking['Sum Trades'].plot(kind='bar', color='skyblue', title='Top 5 Companies by Sum of Trades Since Inception')
plt.ylabel('Sum of Trades')
plt.xlabel('Symbol')
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(Nifty_50.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Strong correlations appear in red indicating relationships between variables, while weaker correlations are show in blue

```python
# Box Plot to Check Outliers in Trades per Symbol
plt.figure(figsize=(10, 6))
sns.boxplot(x='Symbol', y='Trades', data=Nifty_50)
plt.title('Box Plot of Trades for Each Symbol')
plt.xticks(rotation=90)
plt.show()
```

Stocks like RELIANCE, SBIN, and HDFCBANK, exhibit a wider spread of trades suggesting more variability in their trading volumes over time.

```python
# Time-Series Analysis
if 'Date' in Nifty_50.columns:
    Nifty_50['Date'] = pd.to_datetime(Nifty_50['Date'])  # Ensure Date is in datetime format
    trades_per_date = Nifty_50.groupby('Date')['Trades'].sum()

    plt.figure(figsize=(12, 6))
    plt.plot(trades_per_date)
    plt.title('Total Trades Over Time')
    plt.xlabel('Date')
    plt.ylabel('Number of Trades')
    plt.grid(True)
    plt.show()
```

```python
# Analyzing Trades for a Specific Symbol (e.g., 'RELIANCE')
reliance_data = Nifty_50[Nifty_50['Symbol'] == 'RELIANCE']
plt.figure(figsize=(10, 6))
plt.plot(reliance_data['Date'], reliance_data['Trades'], label='RELIANCE Trades')
plt.title('Trades Over Time for RELIANCE')
plt.xlabel('Date')
plt.ylabel('Number of Trades')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
